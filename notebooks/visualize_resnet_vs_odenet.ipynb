```python
import pickle
from matplotlib import pyplot as plt
```

```python
with open('resnet.pkl', 'rb') as f:
    resnet_data = pickle.load(f)
    
with open("odenet.pkl", 'rb') as f:
    odenet_data = pickle.load(f)
```

```python
plt.plot(resnet_data[0].values(), label='resnet_train_loss')
plt.plot(odenet_data[0].values(), label='odenet_train_loss')
plt.legend(loc="upper left")
plt.show()
```

```python
plt.plot(resnet_data[1].keys(), resnet_data[1].values(), label='resnet_test_loss')
plt.plot(odenet_data[1].keys(), odenet_data[1].values(), label='odenet_test_loss')
plt.legend(loc="upper left")
plt.show()
```

```python
plt.plot(resnet_data[2].keys(), resnet_data[2].values(), label='resnet_test_acc')
plt.plot(odenet_data[2].keys(), odenet_data[2].values(), label='odenet_test_acc')
plt.legend(loc="upper left")
plt.show()
```
